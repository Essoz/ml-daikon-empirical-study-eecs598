```python
import json
import matplotlib.pyplot as plt
```

```python
res = json.load(open('./case_3_res.json', 'r'))
```

```python
res
# plot the loss and validation loss with the number of epochs
loss = [r['loss'] for r in res]
val_loss = [r['valid_loss'] for r in res]
plt.plot(loss)
plt.plot(val_loss)
plt.legend(['loss', 'val_loss'])
plt.xlabel('epoch')
plt.ylabel('loss')
plt.title('Loss curves of the original pipeline')
plt.show()

# plot the accuracy and validation accuracy with the number of epochs
acc = [r['Valid_Accuracy'] / 100 for r in res]
plt.plot(acc)
plt.ylabel('validation accuracy')
plt.xlabel('epoch')
plt.title('Validation Accuracy of the original pipeline')
```
